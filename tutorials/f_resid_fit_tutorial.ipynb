# How to Use the Residual Frequency Fit GUI

The FreqOffsetFit class in the processing pipeline, unless you explicitly say otherwise, will open a GUI window in which you will make a polynomial fit to residual frequency. Some of the time, the default fit looks great and you don't need to do any adjusting (like in rev7E). However, in most cases, you will need to do some adjusting, hence why the GUI exists. Since the default fit almost always needs adjusting, we highly recommend you don't turn the GUI off if you're running an occultation for the first time!

To follow along with this example, go to the "pipeline" directory in rss_ringoccs, and type in:

```
python f_resid_fit_tutorial_rev9E_X14_e2e.py
```

This will take a few minutes to get to the GUI - depending on your computer, it'll be somewhere between 5 and 15 minutes.

## Zooming In on the y-axis

Once the time-consuming parts are over, a window will pop up that looks like this:

![title](figs/f_resid_fit_tutorial_image1.png)

This is much more zoomed out than what we want, which is typically how the residual frequency first appears on the screen. To fix this, click the magnifying glass on the bottom-left of the window to zoom in on the y-axis:

![title](figs/f_resid_fit_tutorial_image2.png)

Continue this process until the y-range is about -0.8 Hz to +1 Hz. When you get to that point, here's what the window should look like:

![title](figs/f_resid_fit_tutorial_image3.png)

## Refining the Free-space Regions

Once you're zoomed in like this, the free-space regions where the signal has little to no ring material to pass through are, in terms of seconds past midnight (SPM): 1) a little before 64000 to a little after 6600, and 2) a little before 68000 to a little past 74000. In this case, most of this default fit actually looks pretty good. However, it doesn't capture the very beginning (near 64000 SPM) and very end (near 74000 SPM) of free-space regions so well. Using the GUI, we can improve the fit by adjusting the polynomial fit order and the free-space regions used in the fit.

### Start of the First Free-space Region

To refine the free-space regions used, we put our entry into the text box at the top right labeled "Fit range (SPM)". Before entry, we want to zoom in on each free-space edge mentioned in the last paragraph to find a more exact number for it. Starting off with the first edge just before 64000 SPM, we click and drag to zoom in around this region:

![title](figs/f_resid_fit_tutorial_image4.png)

After you zoom in around here, we can see more clearly that the start of this free-space region is about 63700 SPM, which we will enter into the text box and then click the back arrow to go back to the previous view:

![title](figs/f_resid_fit_tutorial_image5.png)

### End of the First Free-space Region

After going back to the previous view, we'll similarly zoom in on the edge of this free-space region, seen to be just after 66000 SPM on figure 3. Once in this new view, the end of this free-space region is about 66300. When we enter this into the text box, we separate it from the start of the free-space region using a comma:

![title](figs/f_resid_fit_tutorial_image6.png)

### Start and End of the Second Free-space Region

When we click the back arrow, we once again return to the same view as figure 3. We've completed the first free-space region, and now continue to the second free-space region. Referring to figure 3, we zoom in on the edge of the free-space region just before 68000 SPM. In this zoomed-in view, the edge looks like it's at about 67200. Since this is a new free-space region, we need to separate it from the previous one with a semicolon:

![title](figs/f_resid_fit_tutorial_image7.png)

Repeating a similar process, the end of the second free-space region is at about 74300 SPM, which we separate from the beginning (at 67200 SPM) with a comma.

## Setting the Entered Free-space Regions

All taken into account, our text box entry looks like this:

```
63700, 66300 ; 67200, 74300
```
When you hit the "Set fit range (SPM)" button, the routine gets a fit using these input free-space regions. When you press the button, the first view that pops up probably won't clearly show an improved fit very much:

![title](figs/f_resid_fit_tutorial_image8.png)

When we zoom in, the edges of the free-space regions are captured a little better by the fit:

![title](figs/f_resid_fit_tutorial_image9.png)

## Adjusting Order of Polynomial Fit

There's still a few wiggles not quite captured by the fit, so we can increase the fit order using the "Fit order" box on the top left:

![title](figs/f_resid_fit_tutorial_image10.png)

## Finishing the GUI

When you're satisfied with the fit you've made, just click the "OK" button. Another window will come up asking you if you're sure, and if you are, then click "Yes".
